```python
import json

with open("C:\\Users\HP\\\GUVI_PROJ\\sample_airbnb.json",'r') as json_data:
    data=json.load(json_data)

import pandas as pd
normalized_data=pd.json_normalize(data)
df=pd.DataFrame(normalized_data)
print(df['_id'])
```

```python
unhashable_cols = []
for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, list)).any():
        unhashable_cols.append(col)

# Convert unhashable types to hashable (e.g., convert lists to strings)
for col in unhashable_cols:
    df[col] = df[col].apply(lambda x: str(x) if isinstance(x, list) else x)

# Check for duplicates
num_duplicates = len(df) - len(df.drop_duplicates())
print(f"Number of duplicate rows: {num_duplicates}")

```

```python
def is_integer(value):
    try:
        # Check if the value can be converted to an integer
        int(value)
        return True
    except ValueError:
        return False

# Apply the function to the object column to detect integers
df_2['is_integer'] = df_2['_id'].apply(is_integer)

# Display the DataFrame with the detection results
print(df_2)
```

```python
print(df_2.loc[~df_2['is_integer'], '_id'])
```
